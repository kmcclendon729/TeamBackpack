```python
csv_path = "schools.csv"
schools = pd.read_csv(csv_path)
schools.head()
```

```python
csv_path = "output/pubsch_data.csv"
pubsch_data = pd.read_csv(csv_path)
pubsch_data.head()
```

```python
pubsch_data = pubsch_data.drop_duplicates()
```

```python
# skip over this cell. 
#I did use this code to see the columns in the table
#schools.columns
# Removed several columns

schools = schools[["County Code", "District Code", "School Code", "CAASPP Reported Enrollment", "Students Tested", "Percentage Standard Exceeded", "Percentage Standard Met", "Percentage Standard Met and Above", "Percentage Standard Nearly Met", "Percentage Standard Not Met", "Students with Scores"]]
schools.tail()
```

```python
#schools.dtypes
#I will use CDScode as my merging tablet and combine county, district and School code as the make up the CDSCode
schools["CDSCode"]=schools["County Code"].astype(str)+schools["District Code"].astype(str)+schools["School Code"].astype(str)
schools
```

```python
# merge both dataframes
schools = schools.drop_duplicates()
combined_data = schools.merge(pubsch_data, on="CDSCode", how="inner",)
combined_data.head()
```

```python
#rearange the columns again
combined_data = combined_data[["CDSCode","School", "County", "Charter", "Magnet", "EILCode","CAASPP Reported Enrollment","Students Tested", "Percentage Standard Exceeded", "Percentage Standard Met", "Percentage Standard Met and Above", "Percentage Standard Nearly Met", "Percentage Standard Not Met", "Students with Scores" ]]
combined_data.head()
```

```python
combined_data.to_csv("output/combined_data.csv", index=False, header=True)
```

```python
# I want to see the only high schools  in Placer counyty then by type "public vs charter"
Placer_sch_test = combined_data.loc[(combined_data["EILCode"] == "HS") & (combined_data["County"] == "Placer")]
Placer_sch_test.to_csv("output/Placer_sch_test.csv", index=False, header=True)
Placer_sch_test
```
